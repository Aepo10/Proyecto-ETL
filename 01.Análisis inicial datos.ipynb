```python
#Importo pandas para trabajar con los datos
import pandas as pd
#Importo numpy para trabajar con los datos
import numpy as np
#Importo matplot y seaborn para graficar
import matplotlib.pyplot as plt
import seaborn as sns
#Importo pyarrow para trabajar con los datos
import pyarrow.parquet as pq


```

```python
#cargo los datos del archivo parquet
df = pd.read_parquet("reservas_hoteles.parquet", engine='pyarrow')
df.head()
```

```python
#Hago un pequeño analisis de los datos
df.info()
```

```python
#convierto en tipo fechas las columnas con fechas
df["fecha_reserva"]=pd.to_datetime(df["fecha_reserva"])
df["inicio_estancia"]=pd.to_datetime(df["inicio_estancia"])
df["final_estancia"]=pd.to_datetime(df["final_estancia"])
```

```python
#analizo si existen IDs duplicados
df["id_reserva"].duplicated().sum()
```

```python
#existen 98 duplicados, vamos a ver cuales son para analizarlos
df_dup = df[df.duplicated(keep=False)].sort_values("id_reserva")
df_dup
```

```python
#hemos comprobado que existen 98 filas duplicadas exactamente iguales, vamos a eliminarlas
df.drop_duplicates(inplace=True)
#comprobamos que ya no existen duplicados
df[df.duplicated(keep=False)].sort_values("id_reserva")
```

El número de valores únicos de ID_Cliente no es el mismo que el total lo cual nos indica que hay clientes que han hecho varias reservas. 
Sin embargo el valor de valores unicos de clientes es menor que el de valores únicos de mail. Esto carece de sentido, cada mail de cliente debería tener asociado un id_cliente diferente, un mismo cliente no puede tener varios mails.  

```python
#Creamos un nuevo data frame filtrando por los valores únicos de los mails y borramos los duplicado
df_listado_clientes = df[["mail","nombre","apellido"]].drop_duplicates()
#Ahora creamos una nueva columna que nos de un ID unico para cada cliente del 1 al 14.905 usando el indice y concatentando "cl" delante
df_listado_clientes["ID_cliente_correcto"]= "CL" + (df_listado_clientes.index+1).astype(str)
df_listado_clientes.head()
```

```python
df_listado_clientes.describe()
```

```python
#ya tenemos el listado de clientes correcto, ahora vamos a guardarlo en csv
df_listado_clientes.to_csv("listado_clientes.csv",index=False)
```

```python
#Cambiaré el ID de cliente original por el correcto en el data frame original usando una fórmula map
df["id_cliente"]= df["mail"].map(df_listado_clientes.set_index("mail")["ID_cliente_correcto"])
df.head()
```
