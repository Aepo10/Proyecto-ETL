```python
#Importo pandas para trabajar con los datos
import pandas as pd
#Importo numpy para trabajar con los datos
import numpy as np
#Importo matplot y seaborn para graficar
import matplotlib.pyplot as plt
import seaborn as sns
#Importo pyarrow para trabajar con los datos
import pyarrow.parquet as pq


```

```python
#cargo los datos del archivo parquet
df = pd.read_parquet("datos/reservas_hoteles_OK.parquet", engine='pyarrow')
df.head()
```

```python
#Hago un pequeño analisis de los datos
df.info()
```

```python
#convierto en tipo fechas las columnas con fechas
df["fecha_reserva"]=pd.to_datetime(df["fecha_reserva"])
df["inicio_estancia"]=pd.to_datetime(df["inicio_estancia"])
df["final_estancia"]=pd.to_datetime(df["final_estancia"])
```

```python
#analizo si existen IDs duplicados, ya que es la base de la tabla
df["id_reserva"].duplicated().sum()
```

```python
#existen 98 duplicados, vamos a ver cuales son para analizarlos
df_dup = df[df.duplicated(keep=False)].sort_values("id_reserva")
df_dup
```

```python
#hemos comprobado que existen 98 filas duplicadas exactamente iguales, vamos a eliminarlas
df.drop_duplicates(inplace=True)
#comprobamos que ya no existen duplicados
df[df.duplicated(keep=False)].sort_values("id_reserva")
```

```python
df.describe(include=object).T
```

El número de valores únicos de ID_Cliente no es el mismo que el total lo cual nos indica que hay clientes que han hecho varias reservas. 
Sin embargo el valor de valores unicos de clientes es menor que el de valores únicos de mail. Esto carece de sentido, cada mail de cliente debería tener asociado un id_cliente diferente, un mismo cliente no puede tener varios mails.  

```python
#Creamos un nuevo data frame filtrando por los valores únicos de los mails y borramos los duplicado
df_listado_clientes = df[["mail","nombre","apellido"]].drop_duplicates()
#Ahora creamos una nueva columna que nos de un ID unico para cada cliente del 1 al 14.905 usando el indice y concatentando "cl" delante
df_listado_clientes["ID_cliente_correcto"]= "CL" + (df_listado_clientes.index+1).astype(str)
df_listado_clientes.head()
```

```python
df_listado_clientes.describe()
```

```python
#guardo estos datos en un archivo csv en la carpetas datos
df_listado_clientes.to_csv("datos/listado_clientes.csv", index=False)
```

```python
#Cambiaré el ID de cliente original por el correcto en el data frame original usando una fórmula map
df["id_cliente"]= df["mail"].map(df_listado_clientes.set_index("mail")["ID_cliente_correcto"])
df.head()
```

```python
#la columna id reserva ya es correcta. Transformaré el id hotel en objeto para que todas las columnas estén en el tipo correcto
df["id_hotel"]=df["id_hotel"].astype("object")
df.info()
```

```python
#Observamos que existen nulos en las fechas de reserva, precio por noche y estrellas, vamos a filtrar por alguna
df_reservas_nulas = df[df["fecha_reserva"].isnull()]
df_reservas_nulas.describe(include="all").T
```

Todas las fechas de reserva vacias, no tienen ni calificación ni precio voy a analizar un poco más de ellas pero seguramente sean reservas caídas. Analizo cual es el valor único de competencia.

```python
df_reservas_nulas["competencia"].unique()
```

Se trata de reservas de la copetencia, por lo que ya veremos como las rellenamos.

```python
#Filtramos por los valors de competencia que tienen False para ver si seguimos teniendo algun valor nulo
df_propios = df[df["competencia"]==False]
df_propios.info()
```

```python
#Vamos a ver los valores nulos de precio por noche en nuestros hoteles
df_propios[df_propios["precio_noche"].isnull()].describe(include="all").T
```

Observamos que los valores nulos no son demasiados y eliminar las filas nos haría quitar información que puede ser valiosa. Si existen valores vacios en el precio puede ser que haya habido un error en el pago y nos interese seguir investigando. Dejamos los datos debido a ese razonamiento. Los otros datos vacios son inicio de estancia y final de estancia que analizaremos e intentaremos rellenar

```python
#Vamos a ver los valores unicos de inicio y final de estancia
df["inicio_estancia"].unique()
```

```python
df_propios["final_estancia"].unique()
```

```python
#Como todos los valores únicos son el mismo o nulos, vamos a rellenar con el valor único
df["inicio_estancia"] = '2025-03-01 00:00:00'
df["final_estancia"] = '2025-03-02 00:00:00'
df["final_estancia"].unique()
```

```python
#Para crear la base de datos de hoteles, necesitaremos agrupar por nombre de hotel, competencia y haremos una media de las estrellas redondeada a dos decimales
df_propios_def = df_propios.groupby(["nombre_hotel","competencia","ciudad"]).agg({"estrellas":"mean"}).reset_index()
#redondeamos la columna estrellas a dos decimales
df_propios_def["estrellas"]=df_propios_def["estrellas"].round(2)
df_propios_def
```

```python
#Compruebo que cada hotel tenga un solo número de id_hotel y que la competencia sea la misma para todos los hoteles
df.groupby("id_hotel")["nombre_hotel"].nunique()
```

Vemos que hay varios nombres de hoteles que tienen el mismo id, esto no tiene sentido por lo que voy a mapear de nuevo los ids de los hoteles pero una vez sacado los del scrapping de la competencia.

```python
#Vamos a crear la tabla de hoteles agrupando los datos de la competencia y los nuestros y creando un id_hotel único
df_compe_scrappeo= pd.read_csv("datos/hoteles_competencia.csv")
df_compe_scrappeo["id_hotel"] = "H" + (df_compe_scrappeo.index+1).astype(str)
df_compe_scrappeo
```

```python
df.info()
```

```python
#Ahora que tenemos los datos de la competencia vamos a rellenar los datos nulos de estrellas y precio con los datos de la competencia asociando a cada por id de la competencia
#Buscamos los nombres únicos de los hoteles del df que sean competencia true

df_compe_original = df[df["competencia"] == True]
df_compe_original_unicos = pd.DataFrame(df_compe_original["id_hotel"].unique())
df_compe_original_unicos.rename(columns={0:"id_hotel"}, inplace=True)




```

```python
#A cada hotel de la competencia le asignaremo un ID de hotel único OK para sustituirlo con los datos del scrapeo
df_compe_original_unicos["id_hotel_OK"] = "H" + (df_compe_original_unicos.index+1).astype(str)
df_compe_original_unicos.rename(columns={0:"nombre_hotel"}, inplace=True)
df_compe_original_unicos
```

```python
#Sustituimos la columna de ID Hotel por la nueva creada en el df original usando el map
df_reservas_compe = df[df["competencia"]==True]


df_reservas_compe["id_hotel"]=df_reservas_compe["id_hotel"].map(df_compe_original_unicos.set_index("id_hotel")["id_hotel_OK"])
df_reservas_compe
```

```python
df_reservas_compe.columns
```

```python
#Ahora mapeando al ID de hotel correcto, vamos a rellenar los valores nulos de estrellas, nombre y precio por noche usando map
df_reservas_compe['estrellas']=df_reservas_compe["id_hotel"].map(df_compe_scrappeo.set_index("id_hotel")["valoracion"])
df_reservas_compe['precio_noche']=df_reservas_compe["id_hotel"].map(df_compe_scrappeo.set_index("id_hotel")["precio"])
df_reservas_compe['nombre_hotel']=df_reservas_compe["id_hotel"].map(df_compe_scrappeo.set_index("id_hotel")["nombre"])
df_reservas_compe
```

```python
#rellenaremos la columna de fecha reserva con la fecha de hoy para que no haya nulos
df_reservas_compe["fecha_reserva"]="2025-02-20 00:00:00"
```

```python
#A los hoteles propios les asignaremos un ID de hotel único
df_propios_def["id_hotel"] = "H" + (df_propios_def.index+200).astype(str)
df_propios_def
```

```python
#Ahora vincularemos el ID de hotel a las reservas propias
df_reservas_propias = df[df["competencia"]==False]
df_reservas_propias
#Ahora vincularemos el ID de hotel a las reservas propias
df_reservas_propias["id_hotel"]=df_reservas_propias["nombre_hotel"].map(df_propios_def.set_index("nombre_hotel")["id_hotel"])
df_reservas_propias
```

```python

#Ahora unimos el data frame de hoteles de la competencia con el de nuestros hoteles para tener un único data frame de reservas
df_reservas_completo = pd.concat([df_reservas_propias,df_reservas_compe])
df_reservas_completo.info()
```

```python
#Guardamos el data frame en un archivo csv en la carpeta datos
df_reservas_completo.to_csv("datos/reservas_completo.csv", index=False)
```

```python
df_reservas_completo.head()
```

```python
#Hago un dataframe de los datos con el id hotel, nombre de hotel, competencia, Valoración e ID_ciudad con un groupby
df_hoteles_completo = df_reservas_completo.groupby(["id_hotel","nombre_hotel","competencia"]).agg({"estrellas":"mean"}).reset_index()
df_hoteles_completo["estrellas"]=df_hoteles_completo["estrellas"].round(2)
df_hoteles_completo["id_ciudad"]=1
df_hoteles_completo
```

```python
#Convertimos el data frame en un archivo csv en la carpeta datos
df_hoteles_completo.to_csv("datos/hoteles_completo.csv", index=False)
```
